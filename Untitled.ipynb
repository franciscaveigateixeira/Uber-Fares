```python
# Import of Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

warnings.filterwarnings('ignore')

print("Imports completed successfully")
```

```python
# Import of the given Dataset
df = pd.read_csv('uber.csv', index_col=0)

print("The dataset has {} rows and {} columns".format(df.shape[0], df.shape[1]))
print("\nFirst 5 rows:")
print(df.head())

print("\n" + "="*60)
print("DATA TYPES")
print("="*60)
print(df.dtypes)

# Remove 'key' column immediately (it adds no value)
df.drop(columns='key', inplace=True)
print("\n'key' column removed")

# Convert pickup_datetime to datetime
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
print("pickup_datetime converted to datetime")

# Optimize passenger_count data type
df['passenger_count'] = df['passenger_count'].astype('int8')
print("passenger_count converted to int8")

print("\n" + "="*60)
print("MEMORY USAGE AFTER OPTIMIZATION")
print("="*60)
print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
print("="*60)
print("MISSING VALUES ANALYSIS")
print("="*60)

# Check for missing values
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]

if len(missing_counts) > 0:
    print("\nColumns with missing values:")
    for col, count in missing_counts.items():
        pct = (count / len(df)) * 100
        print(f"  {col}: {count} missing ({pct:.4f}%)")
    
    # Show the actual rows with missing values
    print("\nRows with missing values:")
    missing_rows = df[df.isnull().any(axis=1)]
    print(missing_rows)
else:
    print("\nNo missing values found in any column")

# Check for duplicate records
duplicate_count = df.duplicated().sum()
print(f"\nDuplicate records: {duplicate_count}")

if duplicate_count > 0:
    print("Duplicate rows found. Will handle them.")
else:
    print("No duplicate records found.")
```

```python
print("="*60)
print("HANDLING MISSING VALUES")
print("="*60)

# Show the row with missing values before treatment
print("\nRow with missing values BEFORE treatment:")
missing_row = df[df['dropoff_longitude'].isna() | df['dropoff_latitude'].isna()]
print(missing_row)

print(f"\nShape before: {df.shape[0]} rows")

# Since both dropoff coordinates are missing, we cannot calculate distance
# For a trip with passenger_count = 0, this is likely an error or canceled trip
# We will drop this row
df = df.dropna(subset=['dropoff_longitude', 'dropoff_latitude'])

print(f"Shape after dropping missing coordinates: {df.shape[0]} rows")
print(f"Rows removed: 1")

# Verify no missing values remain
print("\nRemaining missing values:")
print(df.isnull().sum().sum())
```

```python
print("="*60)
print("HANDLING ZERO FARE AMOUNTS")
print("="*60)

# First extract pickup_hour from datetime
df['pickup_hour'] = df['pickup_datetime'].dt.hour

# Find rows with fare_amount = 0
zero_fares = df[df['fare_amount'] == 0]
print(f"\nRows with fare_amount = 0: {len(zero_fares)}")

if len(zero_fares) > 0:
    print("\nZero fare rows:")
    print(zero_fares[['fare_amount', 'pickup_latitude', 'pickup_longitude', 
                       'dropoff_latitude', 'dropoff_longitude', 'passenger_count']])
    
    # Calculate distance for all rows
    def haversine_distance(lat1, lon1, lat2, lon2):
        lat1_rad = np.radians(float(lat1))
        lon1_rad = np.radians(float(lon1))
        lat2_rad = np.radians(float(lat2))
        lon2_rad = np.radians(float(lon2))
        
        dlat = lat2_rad - lat1_rad
        dlon = lon2_rad - lon1_rad
        a = np.sin(dlat/2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon/2)**2
        c = 2 * np.arcsin(np.sqrt(a))
        r = 6371
        return c * r
    
    df['distance_km'] = df.apply(
        lambda row: haversine_distance(
            row['pickup_latitude'], row['pickup_longitude'],
            row['dropoff_latitude'], row['dropoff_longitude']
        ), axis=1
    )
    
    print("\nDistance calculated for all rows")
    
    # Classify and handle zero fare rows
    canceled_trips = 0
    imputed_trips = 0
    
    for idx in zero_fares.index:
        row = df.loc[idx]
        # Check if pickup and dropoff coordinates are the same (canceled trip)
        if (abs(row['pickup_latitude'] - row['dropoff_latitude']) < 0.0001 and 
            abs(row['pickup_longitude'] - row['dropoff_longitude']) < 0.0001):
            canceled_trips += 1
            print(f"  Index {idx}: Same coordinates - canceled trip (keeping fare=0)")
        else:
            imputed_trips += 1
            print(f"  Index {idx}: Different coordinates - error, will impute")
            # Impute using median fare per km from similar hour
            similar_trips = df[(df['pickup_hour'] == row['pickup_hour']) & 
                               (df['fare_amount'] > 0) &
                               (df['distance_km'] > 0)]
            if len(similar_trips) > 0:
                fare_per_km = (similar_trips['fare_amount'] / similar_trips['distance_km']).median()
                imputed_fare = row['distance_km'] * fare_per_km
                df.loc[idx, 'fare_amount'] = imputed_fare
                print(f"    Imputed fare: ${imputed_fare:.2f}")
            else:
                # Fallback: use global median fare per km
                valid_trips = df[(df['fare_amount'] > 0) & (df['distance_km'] > 0)]
                fare_per_km = (valid_trips['fare_amount'] / valid_trips['distance_km']).median()
                imputed_fare = row['distance_km'] * fare_per_km
                df.loc[idx, 'fare_amount'] = imputed_fare
                print(f"    Imputed fare (global): ${imputed_fare:.2f}")
    
    print(f"\nCanceled trips (fare kept as 0): {canceled_trips}")
    print(f"Imputed fares: {imputed_trips}")
else:
    print("No zero fare rows found.")
```

```python
print("="*60)
print("HANDLING IMPOSSIBLE COORDINATES")
print("="*60)

# Check for coordinates that are zero (invalid)
invalid_pickup = df[(df['pickup_latitude'] == 0) | (df['pickup_longitude'] == 0)]
invalid_dropoff = df[(df['dropoff_latitude'] == 0) | (df['dropoff_longitude'] == 0)]

print(f"Rows with zero pickup coordinates: {len(invalid_pickup)}")
print(f"Rows with zero dropoff coordinates: {len(invalid_dropoff)}")

if len(invalid_pickup) > 0:
    print("\nInvalid pickup coordinates:")
    print(invalid_pickup[['pickup_latitude', 'pickup_longitude', 'fare_amount']])

if len(invalid_dropoff) > 0:
    print("\nInvalid dropoff coordinates:")
    print(invalid_dropoff[['dropoff_latitude', 'dropoff_longitude', 'fare_amount']])

# Drop rows with zero coordinates (impossible locations)
before_drop = len(df)
df = df[~((df['pickup_latitude'] == 0) | (df['pickup_longitude'] == 0) |
          (df['dropoff_latitude'] == 0) | (df['dropoff_longitude'] == 0))]

print(f"\nDropped {before_drop - len(df)} rows with zero coordinates")
print(f"Shape after dropping: {df.shape[0]} rows")

# Verify index 10649069 was removed
if 10649069 in df.index:
    print("Warning: Index 10649069 still present")
else:
    print("Index 10649069 successfully removed")
```

```python
print("="*60)
print("HANDLING NEGATIVE VALUES")
print("="*60)

# Check for negative fare_amount
negative_fares = df[df['fare_amount'] < 0]
print(f"Rows with negative fare_amount: {len(negative_fares)}")

if len(negative_fares) > 0:
    print("\nNegative fare rows:")
    print(negative_fares[['fare_amount', 'distance_km']])
    df = df[df['fare_amount'] >= 0]
    print(f"\nDropped {len(negative_fares)} rows with negative fares")

# Check for negative passenger_count
negative_passengers = df[df['passenger_count'] < 0]
print(f"\nRows with negative passenger_count: {len(negative_passengers)}")

if len(negative_passengers) > 0:
    print(negative_passengers[['passenger_count']])
    df = df[df['passenger_count'] >= 0]
    print(f"Dropped {len(negative_passengers)} rows with negative passenger counts")

# Check for negative distances (should not happen with haversine)
negative_distances = df[df['distance_km'] < 0]
print(f"\nRows with negative distance_km: {len(negative_distances)}")

print(f"\nShape after handling negatives: {df.shape[0]} rows")
```

```python
print("="*60)
print("FILTERING BY NYC BOUNDARIES")
print("="*60)

# NYC bounds
NYC_LAT_MIN = 40.5
NYC_LAT_MAX = 41.0
NYC_LON_MIN = -74.5
NYC_LON_MAX = -73.5

print(f"NYC boundaries:")
print(f"  Latitude: [{NYC_LAT_MIN}, {NYC_LAT_MAX}]")
print(f"  Longitude: [{NYC_LON_MIN}, {NYC_LON_MAX}]")

before_filter = len(df)

# Filter pickup coordinates
df = df[
    (df['pickup_latitude'].between(NYC_LAT_MIN, NYC_LAT_MAX)) &
    (df['pickup_longitude'].between(NYC_LON_MIN, NYC_LON_MAX)) &
    (df['dropoff_latitude'].between(NYC_LAT_MIN, NYC_LAT_MAX)) &
    (df['dropoff_longitude'].between(NYC_LON_MIN, NYC_LON_MAX))
]

after_filter = len(df)
removed = before_filter - after_filter

print(f"\nRows before filtering: {before_filter:,}")
print(f"Rows after filtering: {after_filter:,}")
print(f"Rows removed: {removed:,} ({removed/before_filter*100:.2f}%)")
```

```python
print("="*60)
print("FILTERING REASONABLE TRIPS")
print("="*60)

before_filter = len(df)

# Distance bounds
MIN_DISTANCE = 0.1  # 100 meters minimum
MAX_DISTANCE = 50   # 50 km maximum

# Fare bounds
MIN_FARE = 1.0      # $1 minimum
MAX_FARE = 200.0    # $200 maximum

# Passenger bounds
MAX_PASSENGERS = 6  # UberX max is 6 passengers

print(f"Filters applied:")
print(f"  Distance: [{MIN_DISTANCE}, {MAX_DISTANCE}] km")
print(f"  Fare: [{MIN_FARE}, {MAX_FARE}] USD")
print(f"  Passenger count: <= {MAX_PASSENGERS}")

# Apply filters
df = df[df['distance_km'] >= MIN_DISTANCE]
df = df[df['distance_km'] <= MAX_DISTANCE]
df = df[df['fare_amount'] >= MIN_FARE]
df = df[df['fare_amount'] <= MAX_FARE]
df = df[df['passenger_count'] <= MAX_PASSENGERS]

after_filter = len(df)
removed = before_filter - after_filter

print(f"\nRows before filtering: {before_filter:,}")
print(f"Rows after filtering: {after_filter:,}")
print(f"Rows removed: {removed:,} ({removed/before_filter*100:.2f}%)")

# Show final statistics
print("\n" + "="*60)
print("FINAL DATA STATISTICS AFTER CLEANING")
print("="*60)

print(f"\nTotal trips: {len(df):,}")

print(f"\nFare Amount ($):")
print(f"  Min: {df['fare_amount'].min():.2f}")
print(f"  Max: {df['fare_amount'].max():.2f}")
print(f"  Mean: {df['fare_amount'].mean():.2f}")
print(f"  Median: {df['fare_amount'].median():.2f}")

print(f"\nDistance (km):")
print(f"  Min: {df['distance_km'].min():.2f}")
print(f"  Max: {df['distance_km'].max():.2f}")
print(f"  Mean: {df['distance_km'].mean():.2f}")
print(f"  Median: {df['distance_km'].median():.2f}")

print(f"\nPassenger Count:")
print(df['passenger_count'].value_counts().sort_index())
```

```python
print("="*60)
print("REMOVING PASSENGER_COUNT = 0")
print("="*60)

before_removal = len(df)

zero_passengers = df[df['passenger_count'] == 0]
print(f"Rows with passenger_count = 0: {len(zero_passengers)}")

if len(zero_passengers) > 0:
    df = df[df['passenger_count'] > 0]
    print(f"Removed {before_removal - len(df)} rows")
    print(f"Shape after removal: {df.shape[0]} rows")

print("\n" + "="*60)
print("FINAL CLEANING SUMMARY")
print("="*60)

print(f"""
Original dataset:        200,000 rows
After missing values:    199,999 rows
After zero coordinates:  196,031 rows
After negative values:   196,014 rows
After NYC filter:        195,643 rows
After reasonable trips:  192,500 rows
After removing passenger_count=0: {df.shape[0]:,} rows

Total rows removed: {200000 - df.shape[0]:,} ({((200000 - df.shape[0])/200000)*100:.2f}%)
""")

print(f"\nFare statistics after full cleaning:")
print(f"  Min: ${df['fare_amount'].min():.2f}")
print(f"  Max: ${df['fare_amount'].max():.2f}")
print(f"  Mean: ${df['fare_amount'].mean():.2f}")
print(f"  Median: ${df['fare_amount'].median():.2f}")

print(f"\nDistance statistics after full cleaning:")
print(f"  Min: {df['distance_km'].min():.2f} km")
print(f"  Max: {df['distance_km'].max():.2f} km")
print(f"  Mean: {df['distance_km'].mean():.2f} km")
print(f"  Median: {df['distance_km'].median():.2f} km")
```

```python
print("="*60)
print("FEATURE ENGINEERING - TIME FEATURES")
print("="*60)

# Extract datetime components
df['pickup_hour'] = df['pickup_datetime'].dt.hour
df['pickup_day'] = df['pickup_datetime'].dt.day
df['pickup_month'] = df['pickup_datetime'].dt.month
df['pickup_year'] = df['pickup_datetime'].dt.year
df['pickup_dayofweek'] = df['pickup_datetime'].dt.dayofweek  # Monday=0, Sunday=6
df['pickup_quarter'] = df['pickup_datetime'].dt.quarter

# Cyclical features for hour (important for ML models)
df['hour_sin'] = np.sin(2 * np.pi * df['pickup_hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['pickup_hour'] / 24)

# Cyclical features for day of week
df['dow_sin'] = np.sin(2 * np.pi * df['pickup_dayofweek'] / 7)
df['dow_cos'] = np.cos(2 * np.pi * df['pickup_dayofweek'] / 7)

# Binary indicators
df['is_weekend'] = (df['pickup_dayofweek'] >= 5).astype(int)
df['is_rush_hour'] = ((df['pickup_hour'] >= 7) & (df['pickup_hour'] <= 9) |
                       (df['pickup_hour'] >= 17) & (df['pickup_hour'] <= 19)).astype(int)
df['is_night'] = ((df['pickup_hour'] >= 22) | (df['pickup_hour'] <= 5)).astype(int)

# Categorical features
df['hour_category'] = pd.cut(df['pickup_hour'], 
                              bins=[0, 6, 12, 18, 24], 
                              labels=['Night', 'Morning', 'Afternoon', 'Evening'],
                              right=False)

df['day_name'] = df['pickup_dayofweek'].map({
    0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday',
    4: 'Friday', 5: 'Saturday', 6: 'Sunday'
})

df['month_name'] = df['pickup_month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})

# Season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['season'] = df['pickup_month'].apply(get_season)

print("Time features created:")
print(f"  pickup_hour: {df['pickup_hour'].nunique()} unique values")
print(f"  pickup_dayofweek: {df['pickup_dayofweek'].nunique()} unique values")
print(f"  is_weekend: {df['is_weekend'].sum()} weekend trips")
print(f"  is_rush_hour: {df['is_rush_hour'].sum()} rush hour trips")
print(f"  is_night: {df['is_night'].sum()} night trips")
print(f"  hour_category: {df['hour_category'].unique()}")
print(f"  season: {df['season'].unique()}")

df.head()
```

```python
print("="*60)
print("FEATURE ENGINEERING - INTERACTION FEATURES")
print("="*60)

# Fare per kilometer (rate)
df['fare_per_km'] = df['fare_amount'] / df['distance_km']

# Distance-based interactions
df['distance_night'] = df['distance_km'] * df['is_night']
df['distance_rush'] = df['distance_km'] * df['is_rush_hour']
df['distance_weekend'] = df['distance_km'] * df['is_weekend']

# Hour interactions
df['hour_weekend'] = df['pickup_hour'] * df['is_weekend']
df['hour_rush'] = df['pickup_hour'] * df['is_rush_hour']

# Passenger interactions
df['fare_per_passenger'] = df['fare_amount'] / df['passenger_count']

print("Interaction features created:")
print(f"  fare_per_km: mean = ${df['fare_per_km'].mean():.2f}/km, median = ${df['fare_per_km'].median():.2f}/km")
print(f"  distance_night: mean = {df['distance_night'].mean():.2f} km")
print(f"  distance_rush: mean = {df['distance_rush'].mean():.2f} km")
print(f"  distance_weekend: mean = {df['distance_weekend'].mean():.2f} km")
print(f"  fare_per_passenger: mean = ${df['fare_per_passenger'].mean():.2f}")

# Check for infinite or NaN values
print("\nChecking for issues:")
print(f"  Infinite values in fare_per_km: {np.isinf(df['fare_per_km']).sum()}")
print(f"  NaN values in fare_per_km: {df['fare_per_km'].isna().sum()}")

# Remove any infinite values
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()

print(f"\nShape after removing infinite values: {df.shape[0]} rows")

print("\nFinal dataset columns:")
print(df.columns.tolist())
```

```python
print("="*60)
print("FINAL DATASET SUMMARY")
print("="*60)

print(f"\nTotal rows: {len(df):,}")
print(f"Total columns: {len(df.columns)}")

print("\n" + "="*60)
print("DATA TYPES")
print("="*60)
print(df.dtypes.value_counts())

print("\n" + "="*60)
print("NUMERICAL FEATURES STATISTICS")
print("="*60)
numerical_cols = df.select_dtypes(include=[np.number]).columns
print(df[numerical_cols].describe())

print("\n" + "="*60)
print("CATEGORICAL FEATURES SAMPLE")
print("="*60)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(f"  Unique values: {df[col].nunique()}")
    print(f"  Sample: {df[col].iloc[:3].tolist()}")

# Export to CSV
print("\n" + "="*60)
print("EXPORTING DATASET")
print("="*60)

df.to_csv('uber_cleaned.csv', index=True)
print("Dataset exported to 'uber_cleaned.csv'")

# Also export a smaller sample for quick testing
df_sample = df.sample(n=10000, random_state=42)
df_sample.to_csv('uber_cleaned_sample.csv', index=True)
print("Sample (10,000 rows) exported to 'uber_cleaned_sample.csv'")

print("\n" + "="*60)
print("CLEANING AND FEATURE ENGINEERING COMPLETE")
print("="*60)
print(f"Final shape: {df.shape[0]:,} rows, {df.shape[1]} columns")
print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

```python
print("="*60)
print("CLUSTERING ANALYSIS - PREPARATION")
print("="*60)

# Select features for clustering
clustering_features = ['distance_km', 'pickup_hour', 'fare_per_km', 
                       'is_weekend', 'is_rush_hour', 'passenger_count']

# Prepare data
X = df[clustering_features].copy()

# Handle any infinite values
X = X.replace([np.inf, -np.inf], np.nan)
X = X.dropna()

print(f"Rows for clustering: {len(X):,}")
print(f"Features used: {clustering_features}")

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nFeatures scaled successfully")
print(f"Scaled data shape: {X_scaled.shape}")
```

```python
print("="*60)
print("FINDING OPTIMAL NUMBER OF CLUSTERS")
print("="*60)

# Test different k values
k_values = range(2, 11)
inertias = []
silhouette_scores = []

# Use a sample for faster computation (30,000 rows)
sample_size = 30000
np.random.seed(42)
sample_indices = np.random.choice(len(X_scaled), sample_size, replace=False)
X_sample = X_scaled[sample_indices]

print(f"Using sample of {sample_size:,} rows for elbow method")

for k in k_values:
    print(f"Testing k={k}...", end=" ")
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_sample)
    inertias.append(kmeans.inertia_)
    score = silhouette_score(X_sample, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"Inertia: {kmeans.inertia_:.0f}, Silhouette: {score:.3f}")

# Plot results
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Elbow method
axes[0].plot(list(k_values), inertias, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Number of Clusters (k)')
axes[0].set_ylabel('Inertia')
axes[0].set_title('Elbow Method for Optimal k', fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Silhouette scores
axes[1].plot(list(k_values), silhouette_scores, 'ro-', linewidth=2, markersize=8)
axes[1].set_xlabel('Number of Clusters (k)')
axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Score by k', fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Mark best k
best_k = k_values[np.argmax(silhouette_scores)]
axes[1].axvline(best_k, color='green', linestyle='--', label=f'Best k={best_k}')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"\nBest number of clusters based on Silhouette score: k={best_k}")
print(f"Best Silhouette score: {max(silhouette_scores):.3f}")
```

```python
print("="*60)
print("APPLYING K-MEANS CLUSTERING")
print("="*60)

# Run K-Means on full dataset with k=8
kmeans = KMeans(n_clusters=8, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

print(f"Clustering completed on {len(df):,} rows")
print(f"Cluster distribution:")

# Show cluster sizes
cluster_sizes = df['cluster'].value_counts().sort_index()
for cluster, size in cluster_sizes.items():
    pct = (size / len(df)) * 100
    print(f"  Cluster {cluster}: {size:,} trips ({pct:.1f}%)")

# Analyze cluster characteristics
print("\n" + "="*60)
print("CLUSTER CHARACTERISTICS")
print("="*60)

cluster_summary = df.groupby('cluster').agg({
    'distance_km': 'mean',
    'fare_amount': 'mean',
    'pickup_hour': 'mean',
    'passenger_count': 'mean',
    'fare_per_km': 'mean',
    'is_weekend': 'mean',
    'is_rush_hour': 'mean'
}).round(2)

print("\nCluster Profiles:")
print(cluster_summary)

# Save cluster labels for later use
df.to_csv('uber_with_clusters.csv', index=True)
print("\nData with cluster labels saved to 'uber_with_clusters.csv'")
```
